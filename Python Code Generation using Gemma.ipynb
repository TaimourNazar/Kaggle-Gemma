# Python Code Generation using Gemma (Fundamental Level)

In notebook we will check the cability of Gemma to generate python code.

We will be using **gemma_instruct_2b_en** variant of Gemma modal.

*Note: First we will use Modal to generate python code, then we will run that code to see how well it works.*

```python
import keras
import keras_nlp

import os

#We have three options to use a backend with Keras "Jax", tensorflow" or "torch", but we will be using Jax in this notebook
os.environ["KERAS_BACKEND"] = "jax"

gemma_lm = keras_nlp.models.GemmaCausalLM.from_preset("gemma_instruct_2b_en")

#Get summary of the modal
gemma_lm.summary()
```

```python
gemma_lm.generate("write code to print hello world in python", max_length=100)
```

```python
#Now lets test the generated code

print("Hello world!")
```

```python
gemma_lm.generate("write code to add two numbers in python", max_length=100)
```

```python
#Now lets test the generated code

num1 = 10
num2 = 20

# Add the two numbers
sum = num1 + num2

# Print the sum
print(sum)
```

```python
gemma_lm.generate("write code to subtract two numbers in python", max_length=100)
```

```python
#Now lets test the generated code

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Subtract the two numbers
difference = num1 - num2

# Print the difference
print("The difference between", num1, "and", num2, "is", difference)
```

```python
gemma_lm.generate("write code to add three numbers in python", max_length=100)
```

```python
#Now lets test the generated code

num1 = 10
num2 = 20
num3 = 30

# Add the numbers together
total = num1 + num2 + num3

# Print the total
print(f"The total is: {total}")
```

```python
gemma_lm.generate("write code to add digits in three indices of array in python", max_length=164)
```

```python
#Now lets test the generated code

arr = [1, 2, 3, 4, 5]

# Create a copy of the array so that we don't modify it
copy_arr = arr.copy()

# Add digits in the first index of the array
copy_arr[0] += 2

# Add digits in the second index of the array
copy_arr[1] += 3

# Add digits in the third index of the array
copy_arr[2] += 4

# Print the modified array
print(copy_arr)
```

```python
gemma_lm.generate("write code to perform bubble sort on an array in python", max_length=200)
```

```python
#Now lets test the generated code

def bubble_sort(nums):
    for i in range(len(nums)):
        for j in range(len(nums) - i - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]

def print_array(nums):
    print(nums)

# Example usage
nums = [5, 2, 8, 3, 1]
print_array(bubble_sort(nums))
```

```python
#Code gives None as output beacuse in Python if a funciton doesn't return anything then it implicity None.
#Lets add a return statement in bubble_sort fucntion to check accuracy of generated code by Gemma

def bubble_sort(nums):
    for i in range(len(nums)):
        for j in range(len(nums) - i - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
    return nums
#Return statment added

def print_array(nums):
    print(nums)

# Example usage
nums = [5, 2, 8, 3, 1]
print_array(bubble_sort(nums))
```
